```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

print("Numpy array:", arr)

print("Array Type:", type(arr))

print( arr * 2)
```

```python
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Numpy array:\n", arr2)

print("array dimension:",arr2.ndim)

print("element at index 0,0:",arr2[0,0])
```

```python
arr3 = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("3D Numpy array:\n", arr3)
print("array dimension:", arr3.ndim)
print("shape of arr3:", arr3.shape)

letter = arr3[1, 0, 1] + arr3[0, 1, 0] + arr2[0 , 2]
print("Accessed element:", letter)
```
