```python
import matplotlib.pyplot as plt 

# We will use a simple dataset where y = 2x. The goal is for the model to find the number 2

# Dataset
X = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Parameters 

w = 0.0 # initial weight
b = 0.0 # initial bias
learning_rate = 0.01
epochs = 1000

lossHistory = []

for epoch in range(epochs):
    total_loss = 0
    grad_w = 0
    grad_b = 0
    n = len(X)
    
    for i in range(n):
        # Prediction: y_hat = wx + b
        y_hat = w * X[i] + b
        
        # Calculate Loss (Mean Squared Error)
        error = y[i] - y_hat
        total_loss += error**2
        
        # Calculate Gradients (Partial Derivatives)
        grad_w += -2 * X[i] * error
        grad_b += -2 * error
        
    # Update Weights (Gradient Descent)
    w -= (grad_w / n) * learning_rate
    b -= (grad_b / n) * learning_rate
    
    lossHistory.append(total_loss / n)

print(f"Final Weight: {w:.4f}, Final Bias: {b:.4f}")
```

```python
plt.plot(range(epochs), lossHistory)
plt.title("Loss vs Epochs")
plt.xlabel("Epoch")
plt.ylabel("Mean Squared Error")
plt.grid(True)
plt.show()
```
